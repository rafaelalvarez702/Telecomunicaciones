```python
import pandas as pd

# Crear un DataFrame de ejemplo (asegúrate de tener datos reales de tus trimestres y accesos)
data = {'Trimestre': ['2023-01-01', '2023-02-01', '2023-03-01'],
        'Accesos_por_cada_100_hogares': [500, 510, 520]}
df = pd.DataFrame(data)

# Convertir la columna de trimestres a tipo de dato fecha
df['Trimestre'] = pd.to_datetime(df['Trimestre'])

# Ordenar el DataFrame por trimestre para asegurar un cálculo correcto
df = df.sort_values(by='Trimestre')

# Calcular el KPI
df['KPI'] = ((df['Accesos_por_cada_100_hogares'].shift(1) - df['Accesos_por_cada_100_hogares']) / df['Accesos_por_cada_100_hogares'].shift(1)) * 100

# Mostrar el DataFrame resultante con el KPI calculado
print(df)
```
